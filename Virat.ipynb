# Virat Kholi's 78 Centuries So far

This project is me performing some exploratory data analysis on a dataset from Kaggle about the cricket player Virat Kholi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Virat-Kohli-International-Cricket-Centuries.csv' 
virat_data = pd.read_csv(file_path)
```

```python
# Remove trailing spaces from column names
virat_data.columns = virat_data.columns.str.strip()
```

```python
# Convert the 'Date' column to datetime format
virat_data['Date'] = pd.to_datetime(virat_data['Date'], errors='coerce')
```

```python
# Remove the '*' from the 'Runs' column and convert it to int
virat_data['Runs'] = virat_data['Runs'].str.replace('*', '').astype(int)
```

```python
# Distribution of scores for Virat Kohli
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.histplot(virat_data['Runs'], bins=30, kde=True)
plt.title('Distribution of Virat Kohli\'s Centuries')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.axvline(virat_data['Runs'].mean(), color='r', linestyle='--', label=f"Mean Score: {virat_data['Runs'].mean():.2f}")
plt.legend()
plt.show()
```

```python
# Centuries scored against each team for Virat Kohli
plt.figure(figsize=(14, 8))
sns.countplot(y='Against', data=virat_data, order = virat_data['Against'].value_counts().index)
plt.title('Number of Centuries Scored by Virat Kohli Against Each Team')
plt.xlabel('Number of Centuries')
plt.ylabel('Team')
plt.show()
```

```python
# Home vs. Away performance for Virat Kohli
home_away_counts_vk = virat_data['Ground'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=home_away_counts_vk.index, y=home_away_counts_vk.values)
plt.title('Virat Kohli: Home vs Away Centuries')
plt.xlabel('Venue')
plt.ylabel('Number of Centuries')
plt.show()
```

```python
# Match outcomes when Virat Kohli scored a century
match_result_counts_vk = virat_data['Result'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=match_result_counts_vk.index, y=match_result_counts_vk.values)
plt.title('Match Results when Virat Kohli Scored a Century')
plt.xlabel('Match Result')
plt.ylabel('Number of Centuries')
plt.show()
```
