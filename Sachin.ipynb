# Sachin Tendulkar's 100 Centuries

This project is me performing some exploratory data analysis on a dataset from Kaggle about the cricket player Sachin Tendulkar

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'Sachin Tendulkar - 100 Centuries.csv'
sachin_data = pd.read_csv(file_path)
```

```python
# Remove trailing spaces from column names
sachin_data.columns = sachin_data.columns.str.strip()
```

```python
# Convert the 'Date' column to datetime format
sachin_data['Date'] = pd.to_datetime(sachin_data['Date'], errors='coerce')
```

```python
# Distribution of scores
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.histplot(sachin_data['Score'], bins=30, kde=True)
plt.title('Distribution of Sachin Tendulkar\'s Centuries')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.axvline(sachin_data['Score'].mean(), color='r', linestyle='--', label=f"Mean Score: {sachin_data['Score'].mean():.2f}")
plt.legend()
plt.show()
```

```python
# Centuries scored against each team
plt.figure(figsize=(14, 8))
sns.countplot(y='Against', data=sachin_data, order = sachin_data['Against'].value_counts().index)
plt.title('Number of Centuries Scored Against Each Team')
plt.xlabel('Number of Centuries')
plt.ylabel('Team')
plt.show()
```

```python
# Home vs. Away performance
home_away_counts = sachin_data['H/A'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=home_away_counts.index, y=home_away_counts.values)
plt.title('Home vs Away Centuries')
plt.xlabel('Venue')
plt.ylabel('Number of Centuries')
plt.show()
```

```python
# Player of the match awards
player_of_match_counts = sachin_data['Player of the match'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=player_of_match_counts.index, y=player_of_match_counts.values)
plt.title('Player of the Match Awards in Centuries')
plt.xlabel('Player of the Match')
plt.ylabel('Number of Centuries')
plt.xticks([0, 1], ['No', 'Yes'])  # Convert boolean to Yes/No for clarity
plt.show()
```

```python
# Match outcomes when Sachin scored a century
match_result_counts = sachin_data['Result'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=match_result_counts.index, y=match_result_counts.values)
plt.title('Match Results when Sachin Scored a Century')
plt.xlabel('Match Result')
plt.ylabel('Number of Centuries')
plt.show()
```
